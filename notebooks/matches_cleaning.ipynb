```python
import pandas as pd

# Charger le CSV original (remplace le chemin si besoin)
df = pd.read_csv("matches.csv")

print("Fichier original charg√© :", df.shape, "lignes et colonnes")

# Colonnes utiles pour ton analyse
cols_to_keep = [
    "id",
    "country_id",
    "league_id",
    "season",
    "stage",
    "date",
    "match_api_id",
    "home_team_api_id",
    "away_team_api_id",
    "home_team_goal",
    "away_team_goal"
]

# V√©rifie que ces colonnes existent bien dans ton CSV
print("Colonnes pr√©sentes dans le fichier :", df.columns.tolist())

# Garde seulement celles dont on a besoin
df = df[cols_to_keep]

# Supprime les lignes o√π il manque trop d'infos (par pr√©caution)
df = df.dropna(subset=["home_team_goal", "away_team_goal"])

# Convertit les colonnes de date au bon format
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# V√©rifie le r√©sultat
print("‚úÖ Fichier nettoy√© :", df.shape)
df.head()
```

```python
# Sauvegarder le CSV nettoy√©
df.to_csv("matches_clean.csv", index=False)

print("üéØ Fichier matches_clean.csv cr√©√© avec succ√®s !")
```
