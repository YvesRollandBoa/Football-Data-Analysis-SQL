```python
import pandas as pd

# Charger le CSV original (remplace le chemin si besoin)
df = pd.read_csv("matches.csv")

print("Fichier original chargé :", df.shape, "lignes et colonnes")

# Colonnes utiles pour ton analyse
cols_to_keep = [
    "id",
    "country_id",
    "league_id",
    "season",
    "stage",
    "date",
    "match_api_id",
    "home_team_api_id",
    "away_team_api_id",
    "home_team_goal",
    "away_team_goal"
]

# Vérifie que ces colonnes existent bien dans ton CSV
print("Colonnes présentes dans le fichier :", df.columns.tolist())

# Garde seulement celles dont on a besoin
df = df[cols_to_keep]

# Supprime les lignes où il manque trop d'infos (par précaution)
df = df.dropna(subset=["home_team_goal", "away_team_goal"])

# Convertit les colonnes de date au bon format
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# Vérifie le résultat
print("✅ Fichier nettoyé :", df.shape)
df.head()
```

```python
# Sauvegarder le CSV nettoyé
df.to_csv("matches_clean.csv", index=False)

print("🎯 Fichier matches_clean.csv créé avec succès !")
```
